```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import mysql.connector
import pandas as pd

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="Kishore2400",
  database="banking"
)

print(mydb)
```

```python
query = "Select * from  banking.Customer"
```

```python
df = pd.read_sql(query, mydb)
```

```python
print(df.head())
```

```python
df.tail()
```

```python
bins = [0,100000, 300000, float('inf')]
labels = ['Low', 'Medium', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
plt.title('Income Band Distribution')
```

```python

## examine the distrubution of unique categories
categorical_columns = ['BRId', 'GenderId','IAId','Amount of Credit Cards',
                         'Nationality','Occupation','Fee Structure','Loyalty Classification',
                         'Properties Owned','Risk Weighting','Income Band']
for col in categorical_columns:
    print('VALUE COUNTS FOR COLUMN:',col)
    print(df[col].value_counts())
```

```python
palette = ["#DE0F23", "#2C0C75"]

for i, predictor in enumerate(categorical_columns):
    plt.figure(i)
    ax = sns.countplot(
        data=df,
        x=predictor,
        hue='GenderId',
        palette=palette
    )

    # Add values on each bar
    for container in ax.containers:
        ax.bar_label(container, padding=2,size=6)

    plt.title(f'Distribution of {predictor} by Gender')
    plt.tight_layout()
    plt.show()
```

```python
palette = ["#4A36BB", "#16B75E","#42650C","#C11AAE","#57550F"]

for i, predictor in enumerate(categorical_columns):
    plt.figure(i)
    ax = sns.countplot(
        data=df,
        x=predictor,
        hue='Nationality',
        palette=palette
    )

    # Add values on each bar
    for container in ax.containers:
        ax.bar_label(container, padding=2,size=6)

    plt.title(f'Distribution of {predictor} by Nationality')
    plt.tight_layout()
    plt.show()
```

```python
Numerical_columns = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

plt.figure(figsize=(15,10))
for i, col in enumerate(Numerical_columns):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.show()
```

```python
correlation_matrix = df[Numerical_columns].corr()
plt.figure(figsize=(10,8))
plt.title('Correlation Matrix of Numerical Features')
sns.heatmap(correlation_matrix, annot=True,cmap='coolwarm',fmt=".2f")
plt.show()
```
